# Chainlearn

Some examples on the things you can do...
Suppose you have some high-dimensional data and want to quickly inspect it by a typical PCA-TSNE projection, followed by a quick KMeans. You would have to do something like the

```python
%matplotlib inline
import seaborn as sns
from matplotlib import pyplot as plt
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE 
from sklearn.cluster import KMeans
 
iris = sns.load_dataset('iris')
iris_no_species = iris.drop('species', axis=1)
 
pca = PCA(n_components=3)
tsne = TSNE(n_components=2)

kmeans = KMeans(n_clusters=2)

cluster_labels = kmeans.fit_predict(iris_no_species)

transformed = tsne.fit_transform(pca.fit_transform(iris_no_species))

plt.scatter(transformed[:, 0], transformed[:, 1], c=cluster_labels)
```

```python
import lib.chainlearn
(iris
 .assign(
     species=lambda df: df['species'].LabelEncoder()
 )
 .Lasso(alpha=0.01, target='species')
 .explain()
 .plot
 .bar()
);
```

```python
(iris
 .assign(
     species=lambda df: df['species'].LabelEncoder()
 )
 .RandomForestClassifier(
     n_estimators=100, 
     target='species'
 )
 .explain()
 .plot
 .bar()
);
```

```python
(iris
 .assign(
     species=lambda df: df['species'].LabelEncoder()
 )
 .RandomForestClassifier(
     n_estimators=100,
     target='species'
 )
 .cross_validate(folds=5, scoring='f1_macro')
 .plot
 .hist()
);
```

```python
(iris
 .assign(
     species=lambda df: df['species'].LabelEncoder()
 )
 .RandomForestClassifier(
     n_estimators=100,
     target='species'
 )
 .rename(columns={0: 'label'})
 .plot
 .hist()
)
```

```python
(iris
 .drop('species', axis=1)
 .PCA(n_components=3)
 .TSNE(n_components=2)
 .assign(
     cluster=lambda df: df.KMeans(n_clusters=2)
 )
 .plot
 .scatter(
     x=0, 
     y=1, 
     c='cluster',
     cmap=plt.get_cmap('viridis')
 )
);
```
